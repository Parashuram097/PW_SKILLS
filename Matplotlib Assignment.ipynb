```python
"""
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
"""


import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

```python
"""
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:


import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

"""

data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
]

fig, axs = plt.subplots(2, 2)
for i, ax in enumerate(axs.flatten()):
    x, y = data[i]
    ax.plot(x, y)
    ax.set_title(f'Line Plot {i+1}')

plt.tight_layout()
plt.show()
```

```python
"""
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
"""

#A bar plot is a type of chart that represents categorical data with rectangular bars. The length of each bar corresponds to the value or frequency of a specific category.


import numpy as np
company =np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company Profit")
plt.show()

plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profit")
plt.show()
```

```python
"""
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
"""

"""
A box plot, also known as a box-and-whisker plot, is a type of chart that displays the distribution of a dataset. 
It provides a summary of the data's central tendency, spread, and any potential outliers.
"""


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
