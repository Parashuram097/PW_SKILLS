```python
"""
Q1. What is Multithreading in pyton?why is it used? Name the module used to handle threads in python?
"""

#Multithreading in Python refers to the ability of a program to execute multiple threads simultaneously within a single process

import threading

# Define a function to be executed in a thread
def print_numbers():
    for i in range(1, 6):
        print(i)

# Create a thread and start it
thread = threading.Thread(target=print_numbers)
thread.start()

# Main thread continues execution
for letter in ['a', 'b', 'c']:
    print(letter)

# Wait for the thread to complete
thread.join()
```

```python
"""
Q2. Why threading module is used? Write the use of the following functions:
   1.activeCount()
   2.CurrentThread()
   3.enumerate()
"""

"""
The threading module in Python is used to handle threads and provide a high-level interface for working with multithreading. 
It allows you to create, manage, and control threads in a concurrent program. 
The threading module provides various functions and methods to facilitate thread management and synchronization.
"""

def a():
    print('Hi')


#activeCount()
import threading

# Create and start multiple threads
threads = []
for i in range(5):
    thread = threading.Thread(target=a)
    thread.start()
    threads.append(thread)

# Get the count of active threads
count = threading.activeCount()
print("Active threads:", count)


#currentThread()
import threading

def my_function():
    thread = threading.currentThread()
    print("Current thread:", thread.name)

# Create and start a thread
thread = threading.Thread(target=my_function)
thread.start()


#enumerate()
import threading

# Create and start multiple threads
threads = []
for i in range(5):
    thread = threading.Thread(target=a)
    thread.start()
    threads.append(thread)

# Get references to all active threads
all_threads = threading.enumerate()
print("Active threads:")
for thread in all_threads:
    print(thread.name)
```

```python
"""
Q3. Explain the following functions:
    1.run()
    2.start()
    3.join()
    4.isAlive()
"""

#run()
import threading

class MyThread(threading.Thread):
    def run(self):
        print("Thread is running\n")

thread = MyThread()
thread.start()  # This will call the run() method


#start()
import threading

def my_function():
    print("Thread is running\n")

thread = threading.Thread(target=my_function)
thread.start()  # Start the thread


#join()
import threading

def my_function():
    print("Thread is running\n")

thread = threading.Thread(target=my_function)
thread.start()  # Start the thread
thread.join()  # Wait for the thread to complete

print("Thread has finished\n")


#IsAlive()
import threading
import time

def my_function():
    time.sleep(2)  # Simulate some work
    print("Thread has finished")

thread = threading.Thread(target=my_function)
thread.start()  # Start the thread

if thread.is_alive():
    print("Thread is still running")
else:
    print("Thread has completed")
```

```python
"""
Q4. write a python program to create two threads.
Thread one must print the list of squares and thread two must print the list of cubes.
"""

import threading

def print_squares():
    for i in range(1, 6):
        square = i ** 2
        print(f"Square of {i}: {square}")

def print_cubes():
    for i in range(1, 6):
        cube = i ** 3
        print(f"Cube of {i}: {cube}")

# Create the first thread to print squares
thread1 = threading.Thread(target=print_squares)

# Create the second thread to print cubes
thread2 = threading.Thread(target=print_cubes)

# Start both threads
thread1.start()
thread2.start()

# Wait for both threads to complete
thread1.join()
thread2.join()
```

```python
"""
Q5. State the advantages and disadvantages of multithreading.
"""

"""
                                        
                                        
                                        Advantages of Multithreading:

1-Increased Efficiency: Multithreading allows multiple tasks or operations to be executed simultaneously. This can lead to improved performance and efficiency, especially in programs that involve tasks that can run concurrently, such as I/O operations or parallel computations.

2-Responsiveness: Multithreading helps in maintaining the responsiveness of an application or system. By dividing tasks into separate threads, the main thread or user interface thread remains responsive to user interactions while background tasks or computations are being performed concurrently.

3-Resource Sharing: Threads within the same process share the same memory space, allowing for efficient sharing of data and resources. This can lead to faster communication and reduced overhead compared to using inter-process communication mechanisms.

4-Simplified Design: Multithreading allows for a more modular and organized design of complex systems. By dividing tasks into separate threads, each responsible for a specific functionality, the overall program structure becomes more manageable and easier to understand.

                                    
                                    
                                    
                                    Disadvantages of Multithreading:

1-Complexity: Multithreaded programming introduces additional complexity compared to single-threaded programming. Synchronization and coordination between threads need to be carefully managed to avoid race conditions, deadlocks, and other concurrency issues.

2-Debugging and Testing: Detecting and resolving issues in multithreaded programs can be challenging. Race conditions and thread synchronization problems can be difficult to reproduce and debug, making testing and troubleshooting more complex.

3-Increased Memory Usage: Each thread within a process requires its own stack and resources, which can increase the memory usage of a program. Having too many threads or inefficient thread management can lead to excessive memory consumption.

4-Scalability Limitations: While multithreading can improve performance on systems with multiple cores or processors, there is a limit to the scalability of threads. As the number of threads increases, the overhead of thread management and synchronization can outweigh the benefits, leading to diminishing returns or even degraded performance.
"""
```
