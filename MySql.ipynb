```python
"""
Q5. Explain Primary Key and Foreign Key.
"""

"""
                                                Primary Key
A primary key is a column or a combination of columns in a table that uniquely identifies each row. 
It serves as a unique identifier for the records in a table and ensures data integrity by enforcing uniqueness and 
providing a way to access individual records efficiently. 
Each primary key value must be unique and not null.
"""


"""
                                                Foreign Key
A foreign key is a column or a combination of columns in a table that establishes a link between data in two tables. 
It creates a relationship between two tables based on the values in the foreign key column(s) and 
the primary key column(s) of another table. The foreign key in one table references the primary key in another table, 
creating a parent-child relationship between the tables.
"""
```

```python
"""
Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.
"""


import mysql.connector

# Connect to MySQL
conn = mysql.connector.connect(
  host="localhost",
  user="your_username",
  password="your_password",
  database="your_database"
)

cursor = conn.cursor()
query = "SELECT * FROM customers"
cursor.execute(query)
results = cursor.fetchall()
for row in results:
    print(row)

cursor.close()
conn.close()



#The cursor() method creates a cursor object that allows you to execute SQL queries and fetch results.It acts as a handle to interact with the database.


#The execute() method of the cursor is used to execute an SQL query. It takes the SQL query as a parameter and sends it to the database for execution.
```
