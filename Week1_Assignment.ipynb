```python
"""
Q6. Create a list containing 25 int type data. Using for loop and if-else condition print if the element is
divisible by 3 or not.
"""

l1=[i for i in range(1,26)]
for i in l1:
    if i%3==0:
        print("divisible by 3")
    else:
        print("not divisible by 3")
```
