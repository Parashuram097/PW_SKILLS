```python
"""
Q1. How can you create a Bokeh plot using Python code?
"""

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
p = figure()

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p.line(x, y)
p.title.text = "My Plot"
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
show(p)
```

```python
"""
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
"""

#glyphs are the visual elements that are added to a plot to represent the data. Glyphs can be used to create various types of visualizations such as lines, circles, squares, bars, patches, etc. They allow you to display and encode data in different ways.

output_notebook()
p = figure()

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.line(x, y, line_color='blue', line_width=2)


x_circle = [2, 4, 6, 8, 10]
y_circle = [5, 4, 3, 2, 1]
p.circle(x_circle, y_circle, size=10, fill_color='red', line_color='black')

p.title.text = "Glyph Example"
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"

show(p)
```

```python
"""
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
"""

output_notebook()
p = figure()
p = figure(title="My Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p.line(x, y)

p.xaxis.major_label_text_font = "Arial"
p.xaxis.major_label_text_font_size = "12pt"
p.xaxis.major_tick_line_color = "red"


show(p)
```

```python
"""
Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?
"""

from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.layouts import row
from bokeh.io import output_notebook

output_notebook()
p = figure()

source = ColumnDataSource(data=dict(x=x, y=y))
p.line('x', 'y', source=source)



def update_data(x,y):
    source.data = dict(x=x, y=y)

def callback_function():
    update_data()
    
curdoc().add_periodic_callback(callback_function, 1000)
layout = row(p)
curdoc().add_root(layout)



#  !bokeh serve --show Boken Assignment.py  # command to run server
```
